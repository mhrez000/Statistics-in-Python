Measures of Central Tendency:
Mean (Average): The sum of all values divided by the number of values. It represents the typical or central value in the dataset.
Median: The middle value of a dataset when arranged in ascending order. It's less affected by outliers than the mean.
Mode: The most frequently occurring value(s) in the dataset.

```python
mean_age = df['Age'].mean().__round__(2)

median_fare = df['Fare'].median()

mode_class = df['Pclass'].mode()[0]

print("Mean Age:", mean_age)
print("Median Fare:", median_fare)
print("Mode Class:", mode_class)
```

Measures of Spread or Dispersion:
Variance: A measure of how data points deviate from the mean. It quantifies the spread of data.
Standard Deviation: The square root of the variance. It provides a more interpretable measure of data spread.
Range: The difference between the maximum and minimum values in the dataset.

```python
Age_variance = df['Age'].var().__round__(2)

fare_std = df['Fare'].std().__round__(2)

print("Age Variance:", Age_variance) # the output of 201.11 means that ages of the passangers vary but that unit is in years squared which is not very intuitive to understand.
print("Fare Standard Deviation:", fare_std)
```

Skewness and Kurtosis:
Calculate skewness and kurtosis to understand the shape of the distribution for numerical variables like age and fare.

Skewness tells you which way your data is leaning (asymmetric).

Kurtosis tells you how heavy the tails are and how extreme the outliers are compared to a normal distribution
Normal Distribution: A standard normal distribution has a kurtosis of 3.
Excess Kurtosis: Calculated as 
, which sets the normal distribution baseline to 0.
Types of Kurtosis:
Leptokurtic (Kurtosis > 3, Excess > 0): Fat tails, high peak, high outlier probability.
Mesokurtic (Kurtosis = 3, Excess = 0): Similar to normal distribution.
Platykurtic (Kurtosis < 3, Excess < 0): Thin tails, low peak, low outlier probability.

```python
Mean_age = df['Age'].mean().__round__(2)
Median_age = df['Age'].median()
age_skewness = df['Age'].skew()
Age_kurtosis = df['Age'].kurtosis()

Mean_fare = df['Fare'].mean().__round__(2)
Median_fare = df['Fare'].median()
fare_skewness = df['Fare'].skew()
fare_kurtosis = df['Fare'].kurtosis()

print("Mean Age:", Mean_age)
print("Median Age:", Median_age)
print("Age Kurtosis:", Age_kurtosis)
print(f"Age Skewness: {age_skewness}\n")

print("Mean Fare:", Mean_fare)
print("Median Fare:", Median_fare)
print("Fare Kurtosis:", fare_kurtosis)
print("Fare Skewness:", fare_skewness)

"we got a positive skewness of 0.457 which is mild as we saw in the histogram earlier, a positive skewness is when the tail extends tot he right, and mean is greater than median"
"kurtosis is 17.9 which is very high, meaning that most people paid relatively low or moderate fares but a small number paid insanely high fares creating massive outliers as we can see in the histogram below"

plt.hist(df['Fare'], bins= 20, alpha=0.6, color='skyblue', edgecolor='black')
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

Percentiles:
A percentile is a statistical measure indicating the value below which a given percentage of observations in a group of data falls. It ranks data on a scale of 0 to 100, showing a score's position relative to others; for example, the 90th percentile means a value is higher than 90% of the data
Calculate and interpret percentiles to understand where data values fall within the distribution.

```python
min_fare = df['Fare'].min()
Q1_fare = df['Fare'].quantile(0.25)
median_fare = df['Fare'].median()
Q3_fare = df['Fare'].quantile(0.75)
max_fare = df['Fare'].max()

print("Minimum Fare:", min_fare)
print("First Quartile (Q1) for Fare:", Q1_fare)
print("Median for Fare:", median_fare)
print("Third Quartile (Q3) for Fare:", Q3_fare)
print("Maximum Fare:", max_fare)
```

Coefficient of Variation (CV):
The Coefficient of Variation (CV) is a unitless statistical measure calculated as the ratio of the standard deviation to the mean. Expressed as a percentage, it indicates the relative size of variability to the mean, allowing comparison between datasets with different units or scales. A lower CV indicates higher precision and less relative dispersion.
Purpose: Measures relative dispersion, making it ideal for comparing the variability of datasets that have different units (e.g., comparing height in cm to weight in kg) or wildly different means.

```python
cv_age = df['Age'].std() / df['Age'].mean()
cv_fare = df['Fare'].std() / df['Fare'].mean()
print("Coefficient of Variation for Age:", cv_age)
"0.5 cv for the age means that the spread is about 50% of the average"
print("Coefficient of Variation for Fare:", cv_fare)
"1.5 cv for the fare means that the spread is about 150% of the average which is very high and indicates a lot of variability in the fare prices as we saw before"
```
