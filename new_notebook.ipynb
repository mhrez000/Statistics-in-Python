# How to Create a New Notebook in the "statistics in python" Folder

> This notebook explains the steps required to add a new Jupyter notebook page under the specified directory.

## 1. Navigate to Folder in VSCode

: Use the **Explorer** pane on the left-hand side to locate and expand the **`statistics in python`** directory within your workspace.

## 2. Open Jupyter Notebook Interface

: Open the command palette (`Ctrl+Shift+P`), type **"Jupyter: Create New Blank Notebook"** or select the Jupyter extension icon to interact with notebooks.

## 3. Create New Notebook

: Choose **"New Notebook"** from the palette or Jupyter toolbar; a fresh untitled `.ipynb` file will open.

## 4. Save Notebook to `statistics in python` Directory

: Use **File  Save As** and navigate to the `statistics in python` folder, or drag the tab into the folder in Explorer to move it. Save with an appropriate name.

## 5. Verify File Location

: Confirm the new notebook appears in the `statistics in python` folder in Explorer and open it to ensure it loads correctly.
