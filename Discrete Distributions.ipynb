# Discrete Distributions

# Discrete · Binary . Bernoulli Distribution

**The Bernoulli distribution models a single trial with exactly two outcomes: success (1) with probability p, or failure (0) with probability 1−p. It is the atomic building block of all other binary distributions — Binomial is just n independent Bernoulli trials.

P(X=1) = p     P(X=0) = 1−p     (i.e. P(X=k) = p^k · (1−p)^(1−k))
Variance is maximized at p=0.5 (maximum uncertainty) and is zero at p=0 or p=1 (certain outcome).**

*Did a user click an ad? (p = click-through rate). Is an email spam? (1=spam, 0=ham). Did a patient survive? Did a coin land heads? Did a test pass?*

> Bernoulli is the foundation of logistic regression — which models P(y=1|X) as a Bernoulli probability. Every binary classification problem (churn, fraud, spam, click) assumes the target follows a Bernoulli distribution.

```python
from scipy.stats import bernoulli
import numpy as np

# ─── Bernoulli Distribution ──────────────────────────────────────
# Scenario: Email spam filter. P(email is spam) = 0.3
p = 0.3   # 30% of emails are spam
dist = bernoulli(p)

print("═══ Bernoulli(p=0.3) — Is email spam? ═══")
print(f"P(spam=1)    = {dist.pmf(1):.4f}  ← P(X=1) = p")
print(f"P(not spam=0)= {dist.pmf(0):.4f}  ← P(X=0) = 1-p")
print(f"E(X)   = {dist.mean():.4f}  ← p")
print(f"Var(X) = {dist.var():.4f}  ← p*(1-p)")
print(f"Std(X) = {dist.std():.4f}")
print()

# How variance changes with p (uncertainty peaks at p=0.5)
print("─── Variance across different p values ───")
print(f"{'p':<8} {'Var=p(1-p)':<14} {'Uncertainty bar'}")
for p_val in [0.05, 0.1, 0.2, 0.3, 0.5, 0.7, 0.9, 0.95]:
    v = p_val * (1 - p_val)
    bar = '█' * int(v * 100)
    print(f"{p_val:<8.2f} {v:<14.4f} {bar}")

print()
# Simulation: generate 10 random Bernoulli outcomes
np.random.seed(42)
samples = bernoulli.rvs(p=0.3, size=20)
print(f"20 simulated emails (1=spam): {samples.tolist()}")
print(f"Spam count in sample: {samples.sum()} / 20")
print(f"Sample spam rate:     {samples.mean():.2f}  (true p = 0.30)")

# Key property: Bernoulli is the building block of Binomial
print("\n─── Connection to Binomial ───")
print("Sum of n independent Bernoulli(p) trials = Binomial(n, p)")
n_trials = 100
bern_sum = np.sum(bernoulli.rvs(p=0.3, size=(n_trials,10000)), axis=0)
print(f"Sum of 100 Bern(0.3): mean={bern_sum.mean():.2f}, std={bern_sum.std():.2f}")
print(f"Binom(100,0.3) theory: mean={100*0.3:.2f}, std={(100*0.3*0.7)**0.5:.2f} ✓")
```


# Discrete · Count of Successes . Binomial Distribution

Notation
X ~ Bin(n,p)
Parameters
n≥1, p∈[0,1]
E(X)
np
Var(X)
np(1−p)
Support
{0,1,...,n}
**The Binomial distribution counts the number of successes in n independent, identical Bernoulli trials. Each trial has the same probability p of success, and trials don't affect each other.**

*P(X=k) = C(n,k) · p^k · (1−p)^(n−k)     k = 0, 1, ..., n
C(n,k) = n!/(k!(n−k)!) counts the number of ways to arrange k successes among n trials. As n→∞ with np constant, Binomial → Poisson. As n→∞ with p fixed, Binomial → Normal (by CLT).*

> Real-world scenarios: Out of 1000 website visitors, how many click the buy button (p=0.05)? In 20 drug trials, how many patients recover? Out of 50 manufactured chips, how many are defective?

> Binomial underpins A/B testing, quality control (defect rates), and the loss function of logistic regression. When you evaluate a classifier on n test samples, the number of correct predictions follows Binomial(n, accuracy)

```python
from scipy.stats import binom
import numpy as np

# ─── Binomial Distribution ───────────────────────────────────────
# Scenario: A/B test — 1000 visitors, 5% click rate. How many clicks?
n, p = 1000, 0.05
dist = binom(n, p)

print("═══ Binomial(n=1000, p=0.05) — Website clicks ═══")
print(f"E(X)   = np    = {dist.mean():.2f} clicks expected")
print(f"Var(X) = np(1-p) = {dist.var():.2f}")
print(f"Std(X) = {dist.std():.2f} clicks")
print()

# Key probability questions
print("─── Probability Questions ───")
print(f"P(exactly 50 clicks) = {dist.pmf(50):.6f}")
print(f"P(at most 40 clicks) = {dist.cdf(40):.6f}")
print(f"P(more than 60)      = {1-dist.cdf(60):.6f}")
print(f"P(between 45 and 55) = {dist.cdf(55)-dist.cdf(44):.6f}")
print()

# PMF for small n — easier to see the shape
print("═══ Binomial(n=10, p=0.3) — PMF shape ═══")
n2, p2 = 10, 0.3
print(f"{'k':<5} {'P(X=k)':<12} {'Bar (PMF)'}")
for k in range(n2+1):
    pmf_v = binom.pmf(k, n2, p2)
    bar = '▓' * int(pmf_v * 200)
    marker = ' ← mode' if k == int(p2*(n2+1)) else ''
    print(f"{k:<5} {pmf_v:<12.4f} {bar}{marker}")

print()
# Normal approximation (CLT): works when n is large enough
print("─── Normal Approximation to Binomial ───")
print("Rule of thumb: use Normal approx when np ≥ 5 AND n(1-p) ≥ 5")
from scipy.stats import norm
n_big, p_big = 100, 0.4
mu  = n_big * p_big
sig = (n_big * p_big * (1-p_big))**0.5
print(f"Binom(100,0.4) — exact P(X≤35) = {binom.cdf(35,n_big,p_big):.6f}")
print(f"Normal approx  P(Z≤35.5)       = {norm.cdf(35.5,mu,sig):.6f} (continuity correction)")

# Real DS use case: detecting if conversion rate changed
print("\n─── A/B Testing with Binomial ───")
print("Control: 1000 users, 52 clicks (5.2%). Did it improve from baseline 5%?")
obs_clicks = 52; n_users = 1000; p_baseline = 0.05
# P(seeing ≥52 clicks if true rate is 5%) = p-value
p_value = 1 - binom.cdf(obs_clicks-1, n_users, p_baseline)
print(f"p-value = P(X ≥ 52 | p=0.05) = {p_value:.4f}")
print(f"Conclusion: {'Significant improvement!' if p_value < 0.05 else 'Not significant'}")
```


# Discrete · Count in Interval . Poisson Distribution

Notation
X ~ Pois(λ)
Parameter
λ > 0 (rate)
E(X)
λ
Var(X)
λ
Support
{0,1,2,...}
**The Poisson distribution models the count of events occurring at a constant average rate λ in a fixed interval of time or space. Events must be independent and cannot occur simultaneously. Its most remarkable property: mean equals variance (both equal λ).

P(X=k) = (e^−λ · λ^k) / k!     k = 0, 1, 2, ...
Poisson is the limit of Binomial(n, p) as n→∞ and p→0 with λ=np constant. Use Poisson when you have many trials, rare events, and a known average rate.**

> Real-world scenarios: Calls arriving at a call center per minute (λ=3). Server errors per hour (λ=0.5). Customers arriving at checkout per minute (λ=2). Typos per page (λ=0.1). Car accidents at intersection per week.

*In data science: Poisson regression models count outcomes (purchases per session, errors per day, goals per match). The mean=variance property is a diagnostic — if your count data has variance much larger than the mean (overdispersion), Poisson isn't the right model.*

```python
from scipy.stats import poisson
import numpy as np

# ─── Poisson Distribution ────────────────────────────────────────
# Scenario: A support team gets on average 4 tickets per hour.
lam = 4
dist = poisson(lam)

print("═══ Poisson(λ=4) — Support tickets per hour ═══")
print(f"E(X)   = λ = {dist.mean():.2f}  (UNIQUE: mean = variance!)")
print(f"Var(X) = λ = {dist.var():.2f}")
print(f"Std(X) = √λ = {dist.std():.4f}")
print()

# PMF table
print(f"{'k':<5} {'P(X=k)':<12} {'P(X≤k) CDF':<15} {'Bar'}")
print("-" * 60)
for k in range(13):
    pmf_v = dist.pmf(k); cdf_v = dist.cdf(k)
    bar = '█' * int(pmf_v*150)
    marker = ' ← peak' if k == lam-1 or k == lam else ''
    print(f"{k:<5} {pmf_v:<12.4f} {cdf_v:<15.4f} {bar}{marker}")

print()
# Operational questions
print("─── Operational Questions ───")
print(f"P(0 tickets — empty hour)  = {dist.pmf(0):.4f}  = e^(-4)")
print(f"P(overwhelmed — 8+ tickets)= {1-dist.cdf(7):.4f}")
print(f"P(5 to 10 tickets)         = {dist.cdf(10)-dist.cdf(4):.4f}")
print()

# Scaling: Poisson over different time windows
print("─── Scaling the rate (λ scales linearly with time) ───")
rate_per_hour = 4
for label, t in [('15 min',.25),('30 min',.5),('1 hr',1),('2 hr',2),('8 hr shift',8)]:
    lam_t = rate_per_hour * t
    print(f"  {label:<12}: λ={lam_t:5.1f}, E(tickets)={lam_t:.1f}, Std={lam_t**0.5:.2f}")

print()
# Goodness-of-fit: Is our data actually Poisson?
print("─── Check if data is Poisson (mean ≈ variance) ───")
np.random.seed(42)
real_data = poisson.rvs(mu=4, size=1000)
print(f"Data mean     = {real_data.mean():.4f}")
print(f"Data variance = {real_data.var():.4f}")
print(f"Ratio (var/mean): {real_data.var()/real_data.mean():.4f}  ← should be ≈1 for Poisson")
print("If ratio >> 1: OVERDISPERSION → use Negative Binomial instead")
```


# Discrete · Waiting Time . Geometric Distribution

Notation
X ~ Geom(p)
Parameter
p ∈ (0,1]
E(X)
1/p
Var(X)
(1−p)/p²
Support
{1,2,3,...}
**The Geometric distribution counts the number of trials needed to get the first success in a sequence of independent Bernoulli trials. It has the unique memoryless property: no matter how many failures you've already had, the next trial's probability of success is still p.

P(X=k) = (1−p)^(k−1) · p     k = 1, 2, 3, ...
Memoryless: P(X > m+n | X > m) = P(X > n). Past failures give you zero information about when the next success will come.**

> Real-world scenarios: How many cold calls until you make a sale? How many job applications until you get an offer? How many coin flips until heads? How many login attempts until success? Retries until a network packet is delivered.

 *Geometric models churn/retention (how many months until a user cancels), retry logic in distributed systems, and click-through funnels. Its memoryless property makes it the discrete analog of the Exponential distribution.*

    

```python
from scipy.stats import geom
import numpy as np

# ─── Geometric Distribution ──────────────────────────────────────
# Scenario: Sales calls — each call has 10% chance of closing a deal
p = 0.10
dist = geom(p)

print("═══ Geometric(p=0.1) — Calls until first sale ═══")
print(f"E(X)   = 1/p = {dist.mean():.1f} calls expected until first sale")
print(f"Var(X) = (1-p)/p² = {dist.var():.1f}")
print(f"Std(X) = {dist.std():.2f} calls")
print()

# PMF — probability the FIRST sale happens on call k
print(f"{'Call k':<10} {'P(first sale on call k)':<28} {'Bar'}")
for k in [1,2,3,4,5,7,10,15,20,30]:
    pmf_v = dist.pmf(k)
    bar = '█' * int(pmf_v*500)
    print(f"{k:<10} {pmf_v:<28.6f} {bar}")

print()
# Cumulative: what fraction of salespeople close by call k?
print("─── P(first sale within k calls) = CDF ───")
for k in [1,5,10,15,20,30,50]:
    cdf_v = dist.cdf(k)
    bar = '░' * int(cdf_v*40)
    print(f"  ≤{k:<4} calls: {cdf_v:.4f}  {bar}")

print()
# Memoryless property demonstration
print("─── Memoryless Property ───")
print("You've already failed 10 calls. Does that change anything?")
print(f"P(X > 15 | X > 10) = P(X > 5) = {1 - dist.cdf(5):.4f}")
print(f"P(X > 5)  directly  =           {1 - dist.cdf(5):.4f}  ← SAME!")
print("Past failures are irrelevant — each call is still fresh.")
print()

# Comparing sales reps with different close rates
print("─── Impact of skill level on expected calls ───")
print(f"{'Close rate p':<16} {'E(calls/sale)':<18} {'P(sale on call 1)'}")
for p_val in [0.02, 0.05, 0.10, 0.20, 0.30, 0.50]:
    print(f"{p_val:<16.2f} {1/p_val:<18.1f} {p_val:.2f}")

```


# Discrete · k-th Success . Negative Binomial Distribution

Notation
X ~ NB(r,p)
Parameters
r≥1, p∈(0,1]
E(X)
r(1−p)/p
Var(X)
r(1−p)/p²
Support
{0,1,2,...}
**The Negative Binomial generalizes Geometric (which is NB with r=1). It models the number of failures before the r-th success. In data science, it's more commonly used as a flexible count distribution that allows variance > mean (overdispersion) — fixing Poisson's limitation.

P(X=k) = C(k+r−1, k) · (1−p)^k · p^r     k = 0, 1, 2, ...
When Var(X) > E(X) in your count data (e.g., customer purchases have more variance than Poisson predicts), the Negative Binomial is usually the right model.**

*Real-world scenarios: Number of customer service calls before 3rd resolution. Number of failed drug trials before 5th clinical success. In ML: modeling overdispersed counts like social media shares, sales volumes, or word frequencies in NLP.*

*Negative Binomial regression is used when Poisson regression fails due to overdispersion. It's the standard model for word counts in NLP, insurance claims, social media engagement, and any count data where variance exceeds mean.*

```python
from scipy.stats import nbinom, poisson
import numpy as np

# ─── Negative Binomial ───────────────────────────────────────────
# scipy parameterizes as NB(r, p) where:
#   r = number of successes (n in scipy)
#   p = probability of SUCCESS in each trial

# SCENARIO 1: Trials until 3rd sale (close rate = 20%)
r, p = 3, 0.20
dist = nbinom(r, p)   # X = number of FAILURES before r-th success

print("═══ NB(r=3, p=0.2) — Failures before 3rd sale ═══")
print(f"E(failures)   = r(1-p)/p = {dist.mean():.2f}")
print(f"E(total calls)= r/p      = {r/p:.2f}  (failures + successes)")
print(f"Var(X) = r(1-p)/p² = {dist.var():.2f}")
print()

print(f"{'Failures (k)':<15} {'P(X=k)':<14} {'Bar'}")
for k in range(21):
    pmf_v = dist.pmf(k)
    bar = '█' * int(pmf_v * 400)
    print(f"{k:<15} {pmf_v:<14.4f} {bar}")

print()
# SCENARIO 2: NB as overdispersed Poisson replacement
print("═══ NB as Count Model (Overdispersion) ═══")
print("E-commerce: daily orders per customer. True process is overdispersed.")
np.random.seed(42)

# Simulate overdispersed data (mix of different customer types)
low_buyers  = np.random.poisson(lam=1,   size=700)   # casual buyers
high_buyers = np.random.poisson(lam=8,   size=300)   # power buyers
data = np.concatenate([low_buyers, high_buyers])

print(f"Data mean     = {data.mean():.3f}")
print(f"Data variance = {data.var():.3f}")
print(f"Var/Mean ratio= {data.var()/data.mean():.3f}  ← >> 1, so NOT Poisson")
print()

# Fit Negative Binomial to overdispersed data
from scipy.optimize import minimize
mean_d = data.mean(); var_d = data.var()
# Method of moments: p = mean/var, r = mean²/(var-mean)
p_fit = mean_d / var_d
r_fit = mean_d**2 / (var_d - mean_d)
print(f"Fitted NB: r={r_fit:.2f}, p={p_fit:.4f}")
nb_fit = nbinom(r_fit, p_fit)
print(f"NB fitted mean = {nb_fit.mean():.3f}  (data mean = {mean_d:.3f})")
print(f"NB fitted var  = {nb_fit.var():.3f}  (data var  = {var_d:.3f})")
```


# Discrete · Without Replacement . Hypergeometric Distribution
Sampling without replacement from a finite population

Notation
X ~ HG(N,K,n)
Parameters
N,K,n integers
E(X)
nK/N
Var(X)
nK(N-K)(N-n)/N²(N-1)
Support
max(0,n+K-N)..min(n,K)
**The Hypergeometric models drawing n items without replacement from a population of N items containing K "successes". Unlike Binomial (which assumes replacement or infinite population), here each draw changes the probability of the next.**

> P(X=k) = C(K,k)·C(N−K, n−k) / C(N,n)     k = max(0,n+K−N) to min(n,K)
> 
> As N→∞ with K/N=p constant, Hypergeometric → Binomial(n,p). The key difference from Binomial: sampling without replacement, finite population.
> 
> 
> 
> Real-world scenarios: In a deck of 52 cards (K=13 hearts), deal 5. How many hearts? Quality control: from a batch of 100 (K=5 defective), inspect 10. Fisher's exact test for 2×2 contingency tables. Audit sampling.


* Hypergeometric is the exact distribution behind Fisher's exact test — used for testing independence in small contingency tables (where chi-square approximation breaks down). Also used in gene enrichment analysis, recommendation system evaluation, and any finite-population sampling scenario.*


```python
from scipy.stats import hypergeom, binom
import numpy as np

# ─── Hypergeometric Distribution ─────────────────────────────────
# Parameters: N=population, K=successes in pop, n=sample size

# SCENARIO 1: Card dealing — 52 cards, 13 hearts, deal 5
N, K, n = 52, 13, 5
dist = hypergeom(N, K, n)

print("═══ HG(N=52, K=13 hearts, n=5 cards dealt) ═══")
print(f"E(hearts in hand)  = nK/N = {dist.mean():.4f}")
print(f"Var(X)             = {dist.var():.4f}")
print(f"Std(X)             = {dist.std():.4f}")
print()

print(f"{'Hearts (k)':<14} {'P(X=k)':<14} {'Bar'}")
for k in range(6):
    pmf_v = dist.pmf(k)
    bar = '█' * int(pmf_v * 200)
    print(f"{k:<14} {pmf_v:<14.6f} {bar}")
print()

# Probability questions
print("─── Probability Questions ───")
print(f"P(0 hearts — no hearts at all) = {dist.pmf(0):.6f}")
print(f"P(5 hearts — flush of hearts)  = {dist.pmf(5):.6f}")
print(f"P(at least 3 hearts)           = {1-dist.cdf(2):.6f}")
print()

# SCENARIO 2: Quality control audit
print("═══ Quality Control: HG(N=500, K=25 defective, n=30 inspected) ═══")
N2, K2, n2 = 500, 25, 30
dist2 = hypergeom(N2, K2, n2)
print(f"E(defectives found) = {dist2.mean():.4f}")
print(f"P(find 0 defective) = {dist2.pmf(0):.4f}  ← might miss the problem!")
print(f"P(find ≥1 defective)= {1-dist2.pmf(0):.4f}")
print(f"P(find ≥3)          = {1-dist2.cdf(2):.4f}")
print()

# Comparison: Hypergeometric vs Binomial (replacement vs no replacement)
print("─── HG vs Binomial: effect of 'without replacement' ───")
p_defect = K2/N2  # 5% defect rate
binom_dist = binom(n2, p_defect)
print(f"{'k':<6} {'HG P(X=k)':<18} {'Binom P(X=k)'}")
for k in range(6):
    hg_p = dist2.pmf(k); bi_p = binom_dist.pmf(k)
    print(f"{k:<6} {hg_p:<18.6f} {bi_p:.6f}")
print("→ Very similar when N >> n (population >> sample)")
```
